```python
from datetime import datetime as dt
import math
import torch
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from sympy import pretty
from sympy.abc import pi
```

```python
np.random.seed(42)
torch.manual_seed(42)
```

```python
def generate_coords(num_coords):
    min = -1
    max = 1
    
    xes = (max - min)*torch.rand((num_coords)) + min
    yes = (max - min)*torch.rand((num_coords)) + min
    
    return xes, yes
```

```python
distances = []
times = []

fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(20,15), sharex=True, sharey=True)

min = -1.01
max = 1.01

mkr = '.'

foo, bar = generate_coords(10)
axs[0,0].scatter(foo, bar, alpha=0.5, marker=mkr)
axs[0,0].set_xlim([min,max])
axs[0,0].set_ylim([min,max])
circ = plt.Circle((0,0),1, fill=None)
axs[0,0].add_artist(circ)
axs[0,0].set_aspect('equal', adjustable='box')

time_start = dt.now()
in_circ = torch.where(foo.square() + bar.square() <= 1.)[0].shape[0]
time_finish = (dt.now() - time_start).total_seconds()
times.append(time_finish)

piest = 4 * (in_circ / foo.shape[0])
pidist = abs(math.pi - piest)
distances.append(pidist)

axs[0,0].set_title(r"$\pi \approx${}, dist = {:0.5f}" "\n" r"In Circle = {}, $\it n$ = {}, time = {}" \
                   .format(piest, pidist, in_circ, foo.shape[0], time_finish))

foo, bar = generate_coords(100)
axs[0,1].scatter(foo, bar, alpha=0.5, marker=mkr)
axs[0,1].set_xlim([min,max])
axs[0,1].set_ylim([min,max])
circ = plt.Circle((0,0),1, fill=None)
axs[0,1].add_artist(circ)
axs[0,1].set_aspect('equal', adjustable='box')

time_start = dt.now()
in_circ = torch.where(foo.square() + bar.square() <= 1.)[0].shape[0]
time_finish = (dt.now() - time_start).total_seconds()
times.append(time_finish)

piest = 4 * (in_circ / foo.shape[0])
pidist = abs(math.pi - piest)
distances.append(pidist)

axs[0,1].set_title(r"$\pi \approx${}, dist = {:0.5f}" "\n" r"In Circle = {}, $\it n$ = {}, time = {}" \
                   .format(piest, pidist, in_circ, foo.shape[0], time_finish))

foo, bar = generate_coords(1000)
axs[0,2].scatter(foo, bar, alpha=0.5, marker=mkr)
axs[0,2].set_xlim([min,max])
axs[0,2].set_ylim([min,max])
circ = plt.Circle((0,0),1, fill=None)
axs[0,2].add_artist(circ)
axs[0,2].set_aspect('equal', adjustable='box')

time_start = dt.now()
in_circ = torch.where(foo.square() + bar.square() <= 1.)[0].shape[0]
time_finish = (dt.now() - time_start).total_seconds()
times.append(time_finish)

piest = 4 * (in_circ / foo.shape[0])
pidist = abs(math.pi - piest)
distances.append(pidist)

axs[0,2].set_title(r"$\pi \approx${}, dist = {:0.5f}" "\n" r"In Circle = {}, $\it n$ = {}, time = {}" \
                   .format(piest, pidist, in_circ, foo.shape[0], time_finish))

foo, bar = generate_coords(10000)
axs[1,0].scatter(foo, bar, alpha=0.5, marker=mkr)
axs[1,0].set_xlim([min,max])
axs[1,0].set_ylim([min,max])
circ = plt.Circle((0,0),1, fill=None)
axs[1,0].add_artist(circ)
axs[1,0].set_aspect('equal', adjustable='box')

time_start = dt.now()
in_circ = torch.where(foo.square() + bar.square() <= 1.)[0].shape[0]
time_finish = (dt.now() - time_start).total_seconds()
times.append(time_finish)

piest = 4 * (in_circ / foo.shape[0])
pidist = abs(math.pi - piest)
distances.append(pidist)

axs[1,0].set_title(r"$\pi \approx${}, dist = {:0.5f}" "\n" r"In Circle = {}, $\it n$ = {}, time = {}" \
                   .format(piest, pidist, in_circ, foo.shape[0], time_finish))

foo, bar = generate_coords(100000)
axs[1,1].scatter(foo, bar, alpha=0.5, marker=mkr)
axs[1,1].set_xlim([min,max])
axs[1,1].set_ylim([min,max])
circ = plt.Circle((0,0),1, fill=None)
axs[1,1].add_artist(circ)
axs[1,1].set_aspect('equal', adjustable='box')

time_start = dt.now()
in_circ = torch.where(foo.square() + bar.square() <= 1.)[0].shape[0]
time_finish = (dt.now() - time_start).total_seconds()
times.append(time_finish)

piest = 4 * (in_circ / foo.shape[0])
pidist = abs(math.pi - piest)
distances.append(pidist)
axs[1,1].set_title(r"$\pi \approx${}, dist = {:0.5f}" "\n" r"In Circle = {}, $\it n$ = {}, time = {}" \
                   .format(piest, pidist, in_circ, foo.shape[0], time_finish))

foo, bar = generate_coords(1000000)
axs[1,2].scatter(foo, bar, alpha=0.5, marker=mkr)
axs[1,2].set_xlim([min,max])
axs[1,2].set_ylim([min,max])
circ = plt.Circle((0,0),1, fill=None)
axs[1,2].add_artist(circ)
axs[1,2].set_aspect('equal', adjustable='box')

time_start = dt.now()
in_circ = torch.where(foo.square() + bar.square() <= 1.)[0].shape[0]
time_finish = (dt.now() - time_start).total_seconds()
times.append(time_finish)

piest = 4 * (in_circ / foo.shape[0])
pidist = abs(math.pi - piest)
distances.append(pidist)
axs[1,2].set_title(r"$\pi \approx${}, dist = {:0.5f}" "\n" r"In Circle = {}, $\it n$ = {}, time = {}" \
                   .format(piest, pidist, in_circ, foo.shape[0], time_finish))

plt.tight_layout()
plt.suptitle(r"Estimations of $\pi$", fontsize=14, y=1. )

plt.show()
```

```python
fra = 0.4

fig, ax1 = plt.subplots(figsize=(8,5))

xlab = ["10", "100", "1000", "10000", "100000", "1000000"]

smooth_distances = sm.nonparametric.lowess(distances, range(len(distances)), frac=fra)
a1 = ax1.plot(smooth_distances[:,1], color='g', label='distance')

ax2 = ax1.twinx()

smooth_times = sm.nonparametric.lowess(times, range(len(times)), frac=fra)
a2 = ax2.plot(smooth_times[:,1], color='b', label='compute')

axs = a1 + a2
labs = [l.get_label() for l in axs]

ax1.legend(axs, labs, loc=2)
ax1.set_xlabel(r'$\it n$ samples')
ax1.set_ylabel(r'ABS distance to $\pi$')
ax1.set_yticklabels([])
ax1.set_yticks([])

ax2.set_ylabel('Compute Time')
ax2.set_yticklabels([])
ax2.set_yticks([])

fig.tight_layout()

plt.xticks(range(len(distances)), xlab, rotation=75)
plt.title(r'Estimations abs distance from $\pi$ vs required compute time')
plt.show()
```
